```python
#REmoving Outliers for better results

Q1 = dF_Bcancer_1.quantile(0.20)
Q3 = dF_Bcancer_1.quantile(0.80)
inter_quartile_range = Q3 - Q1

# quartiles and interquartile range

dF_Bcancer_2 = dF_Bcancer_1[~((dF_Bcancer_1 < (Q1 - (1.5 * inter_quartile_range))) | (dF_Bcancer_1 > (Q3 + (1.5 * inter_quartile_range)))).any(axis=1)]
dF_Bcancer_1.shape, dF_Bcancer_2.shape
```

```python
#Create x and y variables
x = dF_Bcancer_2.drop('diagnosis', axis=1).to_numpy()
Y = dF_Bcancer_2['diagnosis'].to_numpy()

#Create Train and Test Datasets
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test = train_test_split(x,Y,test_size = 0.3,random_state = 100) 

#Scale the Data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train2 = sc.fit_transform(x_train)
x_test2 = sc.transform(x_test)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
# Defining and fitting a DecisionTreeClassifier instance
# instantiate the DecisionTreeClassifier model with criterion entropy index

clf_entropy = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=50)


# fit the model
clf_entropy.fit(x_train2, Y_train)
```

```python
y_pred_entropy = clf_entropy.predict(x_test2)
```

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion entropy index: {0:0.4f}'. format(accuracy_score(Y_test, y_pred_entropy)))
```

```python
y_pred_train_entropy = clf_entropy.predict(x_train2)

y_pred_train_entropy
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(Y_train, y_pred_train_entropy)))
```

```python
print('Training set score: {:.4f}'.format(clf_entropy.score(x_train2, Y_train)))

print('Test set score: {:.4f}'.format(clf_entropy.score(x_test2, Y_test)))
```
