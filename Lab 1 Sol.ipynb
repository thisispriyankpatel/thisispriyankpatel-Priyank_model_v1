Importing all the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
```

Importing the dataset

```python
columns = ['Class','AGE','SEX','STEROID','ANTIVIRALS','FATIGUE','MALAISE','ANOREXIA','LIVER BIG','LIVER FIRM','SPLEEN PALPABLE','SPIDERS','ASCITES','VARICES','BILIRUBIN','ALK PHOSPHATE','SGOT','ALBUMIN','PROTIME','HISTOLOGY']
pd = pd.read_csv(r"C:\Users\priyank\Downloads\hepatitis.data", sep=',', names=columns, na_values='?', header=None)
pd.head
```

Filling up the null values with means

```python
pd.fillna(pd.mean(numeric_only=True).round(1), inplace=True)
pd.head(10)
```

Splitting the data into train and test data

```python
Y = pd['Class']
X = pd.loc[:, pd.columns != 'Class']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
print(X)
```

```python
Y = np.asarray(Y).astype('float32')
X = np.asarray(X).astype('float32')
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
Y_train = np.asarray(Y_train).reshape(-1,1)
```

```python
model = Sequential()
model.add(Dense(19, input_dim=19, activation='sigmoid'))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(15, activation='sigmoid'))
model.add(Dense(1, activation=None))
```

Compiling the model

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, epochs=100, batch_size=32, verbose=0)
loss, acc = model.evaluate(X_test, Y_test, verbose=0)
print('Test Accuracy: %.3f' % acc)
```

```python
Y_pred = model.predict(X_test)
Y_pred = (Y_pred > 0.5)
print("Accuracy:", accuracy_score(Y_test, Y_pred))
```
